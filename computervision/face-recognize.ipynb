```python
import cv2
import matplotlib.pyplot as plt

rgb_img = cv2.imread(r"C:\Users\anime\Pictures\Camera Roll\WIN_20250624_11_11_52_Pro.jpg")

gray_img = cv2.cvtColor(rgb_img, cv2.COLOR_BGR2GRAY)

edge_dect = cv2.Canny(gray_img, 100, 200)

# Step 4: Display the result using matplotlib
plt.imshow(edge_dect, cmap='gray')
plt.axis('off')
plt.title("Canny Edge Detection")
plt.show()
```

```python
face = r"C:\Users\anime\Downloads\haarcascade_frontalface_default.xml"
model=cv2.CascadeClassifier(face)

img = cv2.imread(r"C:\Users\anime\Pictures\Camera Roll\WIN_20241212_13_29_05_Pro.jpg")

facepoints = model.detectMultiScale(img)
for x , y , w , h in facepoints:
    cv2.rectangle(img , (x,y) , (x+w , y+h) , (123,45,20) , 3)
plt.imshow(img[::,::,::-1])
plt.show()
```

```python
face_detc = r"C:\Users\anime\Downloads\haarcascade_frontalface_default.xml"
cap = cv2.VideoCapture(0)
ret,frame = cap.read()
while ret:
    ret,frame = cap.read()
    print(ret)
    print(frame)

    if not ret:
        break
    model = cv2.CascadeClassifier(face_detc)
    facepoints = model.detectMultiScale(frame)

    for x,y,w,h in facepoints:
        cv2.rectangle(frame , (x,y) , (x+w,y+h) , (0,0,255),3)

    cv2.imshow('My Face ',frame)
    if cv2.waitKey(10)==27:
        break
cv2.destroyAllWindows()
cap.release()
```
