```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
X,y=load_iris(return_X_y = True)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
mod = KNeighborsRegressor()
```

```python
mod=mod.fit(X, y)
```

```python
pred=mod.predict(X)
```

```python
plt.scatter(pred,y)
```
