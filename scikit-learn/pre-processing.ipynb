```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('drawndata1.csv')
df.head(5)
```

```python
X = df[['x','y']].values
y=df['z'].values=='a'
y # y is of boolean like trye or false
x1=X[:,0] # for x
y1=X[:,1] # for y ofthe df
plt.plot(x1,y1)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler,QuantileTransformer
X_new = QuantileTransformer().fit_transform(X)
x1=X_new[:,0] # for x
y1=X_new[:,1] # for y ofthe df
plot_output=plt.scatter(x1,y1,c=y)
```

```python
plot_output.set_alpha(0.7)  # Make points a bit transparent
plot_output.set_edgecolor('black')  # Give points a border
plot_output(scaler=StandardScaler())
```
