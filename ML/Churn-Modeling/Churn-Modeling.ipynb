```python
data1=data.drop(['RowNumber' , 'CustomerId' , 'Surname'] , axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
data1.Geography = LabelEncoder().fit_transform(data1.Geography)
data1.Gender = LabelEncoder().fit_transform(data1.Gender)
```

```python
inputs = data1.drop(['Exited'] , axis=1)
output = data1.Exited
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test  , y_train , y_test = train_test_split(inputs , output , test_size=0.2)
```

# Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train , y_train)
yp=lr.predict(x_test)
from sklearn.metrics import accuracy_score , recall_score , precision_score , f1_score
print(accuracy_score(y_test , yp))
print(recall_score(y_test , yp))
print(precision_score(y_test , yp))
print(f1_score(y_test , yp))
```

# Using KNN

```python


from sklearn.neighbors import KNeighborsClassifier
d=KNeighborsClassifier(n_neighbors=5)
d.fit(x_train , y_train)
yp=d.predict(x_test)
from sklearn.metrics import accuracy_score , recall_score , precision_score , f1_score
print(accuracy_score(y_test , yp))
print(recall_score(y_test , yp))
print(precision_score(y_test , yp))
print(f1_score(y_test , yp))
```

# Using decision tree

```python
from sklearn.tree import DecisionTreeClassifier
d=DecisionTreeClassifier(criterion='entropy')
d.fit(x_train , y_train)
yp=d.predict(x_test)
from sklearn.metrics import accuracy_score , recall_score , precision_score , f1_score
print(accuracy_score(y_test , yp))
print(recall_score(y_test , yp))
print(precision_score(y_test , yp))
print(f1_score(y_test , yp))
```

# Using naive bayes gaussian nb

```python
from sklearn.naive_bayes import GaussianNB
d=GaussianNB()
d.fit(x_train , y_train)
yp=d.predict(x_test)
from sklearn.metrics import accuracy_score , recall_score , precision_score , f1_score
print(accuracy_score(y_test , yp))
print(recall_score(y_test , yp))
print(precision_score(y_test , yp))
print(f1_score(y_test , yp))
```

# Using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
d=RandomForestClassifier( criterion ='entropy')
d.fit(x_train , y_train)
yp=d.predict(x_test)
from sklearn.metrics import accuracy_score , recall_score , precision_score , f1_score
print(accuracy_score(y_test , yp))
print(recall_score(y_test , yp))
print(precision_score(y_test , yp))
print(f1_score(y_test , yp))
```
