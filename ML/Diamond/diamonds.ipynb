```python
data1=data
```

```python
from sklearn.preprocessing import LabelEncoder
data1.clarity=LabelEncoder().fit_transform(data.clarity)
data1.cut=LabelEncoder().fit_transform(data.cut)
data1.color=LabelEncoder().fit_transform(data.color)
```

```python
data1
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(data=data1.corr() , annot=True)
```

```python
inputs = data1.drop(['price','depth' ] , axis=1)
output=data.price

from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(inputs , output , test_size=0.2)
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_test = sc.fit_transform(x_test)
x_train = sc.fit_transform(x_train)


```

# Using linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score
alg=LinearRegression().fit(x_train , y_train)
pred_y=alg.predict(x_test)
print(mean_squared_error(pred_y , y_test))
print(r2_score(pred_y , y_test))
```

# [polynomial regression] 

```python


from sklearn.preprocessing import PolynomialFeatures

poly=PolynomialFeatures(degree=2)

new_x_train = poly.fit_transform(x_train)
new_x_test=poly.fit_transform(x_test)

reg=LinearRegression()
reg.fit(new_x_train , y_train)
pred_y=reg.predict(new_x_test)
print(mean_squared_error(pred_y , y_test))
print(r2_score(pred_y , y_test))
```

```python
from sklearn.tree import DecisionTreeRegressor,plot_tree
alg = DecisionTreeRegressor(max_depth=3)
alg.fit(x_train , y_train)
yp=alg.predict(x_test)
print(mean_squared_error(yp , y_test))
print(r2_score(yp , y_test))
plt.figure(figsize=(15,15))
plot_tree(alg)
plt.show()
```
