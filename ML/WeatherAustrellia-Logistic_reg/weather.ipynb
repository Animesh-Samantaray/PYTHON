```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='darkgrid', palette='Dark2')
plt.figure(figsize=(12,8))
```

```python
data=pd.read_csv('weatherAUS.csv')
```
