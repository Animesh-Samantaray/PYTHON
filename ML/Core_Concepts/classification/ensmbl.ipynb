```python
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.datasets import load_breast_cancer
```

```python
x,y=load_breast_cancer(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
x_train,y_train,x_test,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,VotingClassifier
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
```

```python
estimators=[
    ('c1',LogisticRegression()),
    ('c2',SVC(probability=True)),
    ('c3',DecisionTreeClassifier(max_depth=3,criterion='gini',splitter='best',min_samples_split=6,max_features=5,max_leaf_nodes=10)),
    ('c4',RandomForestClassifier())
]
```

```python
for estimator in estimators:
    scores = cross_val_score(estimator[1], x, y, cv=10, scoring='accuracy')
    print(estimator[0], np.round(np.mean(scores), 2))

```

```python
vc1=VotingClassifier(estimators=estimators,voting='soft')
p = cross_val_score(vc1, x, y, cv=10, scoring='accuracy', error_score='raise')
print(np.round(np.mean(p),2))
```
