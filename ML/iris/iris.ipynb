```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='darkgrid', palette='Dark2')
import warnings 
warnings.filterwarnings('ignore')
```

```python
data1=pd.read_csv('Iris.csv')
data1
data=data1.drop(['Id'] , axis=1)
```

```python
data.columns = ['sl','sw','pl','pw','Specis']
```

```python
inputs = data.drop('Specis' , axis=1)
output = data.Specis
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(inputs , output , test_size=0.2 , random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(x_train , y_train)
reg.score(x_test,y_test)
```
