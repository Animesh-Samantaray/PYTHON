```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set(style='darkgrid', palette='Dark2')
```

```python
df = pd.read_csv('Iris.csv')
data = df.drop(['Species' , 'Id'],axis=1)
data
```

```python
data.columns = ['sl','sw','pl','pw']
data
```

```python
from sklearn.cluster import KMeans

wcss = []

for i in range(2,21):
    km = KMeans(n_clusters=i , init = 'k-means++')
    km.fit(data)
    wcss.append(km.inertia_)


```

```python
sns.lineplot(x=[i for  i in range(2,21)] , y=wcss , marker='o')
plt.xlabel('no of cluster')
plt.xticks([i for  i in range(2,21)])
plt.ylabel('wcss value')
plt.show()
```

```python
km = KMeans(n_clusters=3)
data['label'] = km.fit_predict(data)
data
```

```python
sns.pairplot(data=data, hue='label')
```
