```python
inputs = data.drop(['outcome','skinThickness'] , axis=1)
output=data.outcome
```

```python
# from sklearn.preprocessing import OneHotEncoder
# from sklearn.compose import ColumnTransformer
# ct=ColumnTransformer([('pregnancies' , OneHotEncoder() , [0])] , remainder='passthrough')
# inputs=np.array(ct.fit_transform(inputs) , dtype=str)
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(inputs , output , test_size=0.1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test) 
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression().fit(x_train , y_train)
yp=lg.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test , yp))
```

```python
from sklearn.metrics import  accuracy_score , precision_score , recall_score , f1_score
print(accuracy_score(y_test , yp))
print(precision_score(y_test , yp))
print(recall_score(y_test , yp))
print(f1_score(y_test , yp))
```

```python
from sklearn.neighbors import KNeighborsClassifier
alg=KNeighborsClassifier(n_neighbors=5)
alg.fit(x_train,y_train)
yp=alg.predict(x_test)
from sklearn.metrics import  accuracy_score , precision_score , recall_score , f1_score
print(accuracy_score(y_test , yp))
print(precision_score(y_test , yp))
print(recall_score(y_test , yp))
print(f1_score(y_test , yp))
```

# naive baised algo is used here for more improvement in the model

```python
from sklearn.naive_bayes import GaussianNB
clf=GaussianNB()
clf.fit(x_train , y_train)
```

```python
yp1=clf.predict(x_test)
```

```python
from sklearn.metrics import  accuracy_score , precision_score , recall_score , f1_score
print(accuracy_score(y_test , yp1))
print(precision_score(y_test , yp1))
print(recall_score(y_test , yp1))
print(f1_score(y_test , yp1))
```

# using svm support vector machine

```python
from sklearn import svm
alg=svm.SVC(kernel='rbf' , C=1000 , gamma=0.1)
alg.fit(x_train , y_train)
yp1=alg.predict(x_test)
print(accuracy_score(y_test , yp1))
print(precision_score(y_test , yp1))
print(recall_score(y_test , yp1))
print(f1_score(y_test , yp1))
```
