```python
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv('data.csv')
dataset
```

## Independent varibales are -> country , age , salary
## dependent variable - > phurchase

```python
# dataset.age.dtype='int'
# dataset.salary.dtype='int' --> wrong way
# dataset['age']=dataset['age'].astype(int)
# dataset['salary']=dataset['salary'].astype(int)
x=dataset[['country' , 'age' , 'salary']].values
x
x[:,1:3]=x[:,1:3].astype(float)
```

```python
y=dataset['purchase'].values
y
```

```python
age_mean = np.nanmean(x[:,1])
age_mean
```

```python
salary_mean=np.nanmean(x[:,2])
salary_mean
```

```python
country_mode = dataset['country'].mode().iloc[0]
```

```python
purchase_mode = dataset['purchase'].mode().iloc[0]
```

## set the means to the nan of dataframe 

```python
dataset['age']=dataset['age'].fillna(age_mean)
dataset['salary']=dataset['salary'].fillna(salary_mean)
dataset['country']=dataset['country'].fillna(country_mode)
dataset['purchase']=dataset['purchase'].fillna(purchase_mode)
```

```python
dataset['age']=dataset['age'].astype(int)
dataset['salary']=dataset['salary'].astype(int)
```
