```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(df.drop(['Exited'],axis=1) , df['Exited'] ,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()

model.add(Dense(3 , activation='sigmoid',input_dim=x_train.shape[1]))
model.add(Dense(1, activation='sigmoid',))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=15)
```

```python
def sgmd(v):
    return 1 if v>0.5 else 0
y_pred=model.predict(x_test)
y_ans=np.where(y_pred>0.5,1,0)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_ans)
```
