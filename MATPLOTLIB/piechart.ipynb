# ipl rcb vs csk

```python
data=pd.read_csv('ipl.csv')
mask=(((data.team1=='Royal Challengers Bangalore') & (data.team2=='Chennai Super Kings')) | ((data.team1=='Chennai Super Kings') & (data.team2=='Royal Challengers Bangalore')))
data[mask]
rcb=data[(mask) & (data.winner=='Royal Challengers Bangalore')].shape[0]
csk=data[(mask) & (data.winner=='Chennai Super Kings')].shape[0]
plt.pie([rcb,csk],labels=['rcb','csk'],colors=['r','y'],autopct='%.3f')
plt.legend()
plt.show()
```

```python
mask1=(data.toss_winner==data.winner)
mask2=(data.toss_winner!=data.winner)
correct = data[mask1].shape[0]
incorrect = data[mask2].shape[0]
l=['toss win->match win','toss loss->match loss']
plt.pie([correct,incorrect],labels=l,colors=['g','r'],shadow=True,autopct='%.3f')
plt.legend()
plt.show()
print(correct,incorrect)
```
