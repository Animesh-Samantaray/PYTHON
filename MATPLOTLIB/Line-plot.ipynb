```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('money.csv')
plt.plot(data.year, data.india,'b', label='India (INR)',)
# plt.plot(data.year, data.england, label='England (GBP)')
plt.plot(data.year, data.japan, 'r',label='Japan (JPY)')
plt.plot(data.year, data.pakistan, label='Pakistan (PKR)')
plt.xlabel('Year')
plt.ylabel('Exchange Rate (1 USD to local currency)')
plt.title('USD Exchange Rate Trends (1975–2024)')
plt.legend()
# plt.xticks(data.year)
plt.figure(figsize=(8,5))
plt.show()
```

```python

for i in data.iloc[: , 1:]:
    plt.plot(data['year'] , data[i] , label=f'{i} money')
plt.savefig('currency.png',dpi=1000)
plt.show()
```
