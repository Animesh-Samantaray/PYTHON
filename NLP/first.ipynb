```python
from nltk.corpus import stopwords
text = "Virat is playing incredibly well in the IPL 2025! ' 😤 🔥 And he is the best batsman in the world and also my favourite "

words = word_tokenize(text.lower())
words

stop_words = set(stopwords.words('english'))
```

```python
cleaned = []
for i in words:
    if i.isalpha() and i not in stop_words:
        cleaned.append(i)

print(cleaned)
```

# training with pos - (parts of speech)

```python
import nltk
nltk.download('averaged_perceptron_tagger')
text = "romeo loves Juliet and Juliet too . They both loves each other like  cloud and rain . everytime they stay at problem places they got married at egland and then travelled in england and returned to india to visit chaturdham "

tokens = word_tokenize(text)
tags = nltk.pos_tag(tokens)

print(tags)
```

```python
import string
text = "romeo loves ......Juliet and Juliet too ['!@#$%^] . They both loves each other like  cloud and rain . everytime they stay at problem places they got married at egland and then travelled in england and returned to india to visit chaturdham "

no_puncutation = [i for i in text if  i not in  string.punctuation ]
no_puncutation = ''.join(no_puncutation)
print(no_puncutation)
```

```python
cleared_data = [i for i in no_puncutation.split() if i.lower() not in stop_words]
cleared_data
```
