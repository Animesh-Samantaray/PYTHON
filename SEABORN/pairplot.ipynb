```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=sns.load_dataset('tips')
sns.pairplot(data,hue='sex',palette='Dark2',kind='kde')## scatter kde reg hist
plt.show()
data
```

```python
sns.pairplot(data,vars=['tip','total_bill'],hue='sex',palette='Dark2')
plt.show()
```

## customize x and y axis

```python
sns.pairplot(data, 
             x_vars=['tip', 'total_bill'], 
             y_vars=['tip', 'total_bill'], 
             hue='sex', 
             palette='Dark2', 
             kind='kde')
```

```python
data=sns.load_dataset('tips')
sns.pairplot(data.head(50),hue='sex',hue_order=['Female','Male'],markers=['<','*'])## scatter kde reg hist
plt.show()
```
