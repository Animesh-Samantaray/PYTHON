# Line Plotting

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
var1=[1,2,3,4,5,6]
var2=[100,300,200,900,9,100]
plt.plot(var1,var2)
plt.show()
```

# unusing seaborn

```python
df=pd.DataFrame({'var1':var1,'var2':var2})
sns.lineplot(x='var1',y='var2',data=df)
plt.show()
```

```python
gl=sns.load_dataset('geyser')
gl
sns.lineplot(x='duration',y='waiting',hue='kind',data=gl.head(30),style='kind',markers=['o','>'])
plt.grid(True)
plt.title('Animesh')
plt.show()
```
