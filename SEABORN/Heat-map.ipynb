```python
# dt=np.linspace(st,end,number-of-data)
dt=np.linspace(1,10,20).reshape(4,5)
# dt=np.floor(dt)
sns.heatmap(dt)
plt.show()
```

```python
dt=sns.load_dataset('anagrams')
dt
```

```python
x=dt.drop(columns=['attnr'] , axis=1)
x
```

```python
sns.heatmap(x.head(7),vmin=1,vmax=7,cmap='PuOr',annot=True)
plt.show()
```

```python
arr=np.array([['a1','a2','a3','a4'],['b1','b2','b3','b4']])
main=np.linspace(1,8,8).reshape(2,4)
seter={
    'fontsize':30,
    'color':'r'
}
sns.heatmap(main,cmap='PuOr',fmt='s',annot=arr,annot_kws=seter,linewidths=10,linecolor='b')
plt.show()
```
