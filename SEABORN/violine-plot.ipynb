```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=sns.load_dataset('tips')
data
```

```python
sns.violinplot(x=data.day,y=data.total_bill,palette='Dark2',hue=data.day , linewidth=2)
plt.show()
```

```python
sns.violinplot(x=data.time , y = data.total_bill , palette='Dark2',hue=data.sex,order=['Lunch','Dinner'])
```

## if we want to spit the hue data

```python
sns.violinplot(x=data.time , y = data.total_bill , palette='Dark2',hue=data.sex,order=['Lunch','Dinner'],split=True,inner='quart')
```

```python
sns.violinplot(x=data.day , y = data.total_bill , palette='Dark2',hue=data.sex,split=True)
```
