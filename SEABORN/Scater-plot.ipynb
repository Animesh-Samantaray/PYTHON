# Scatter ploting

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style("whitegrid")
dt=sns.load_dataset('penguins')
sns.scatterplot(x=dt.bill_length_mm , y=dt.bill_depth_mm,hue=dt.sex,palette=['red','blue'],sizes=(10,30),size=dt.sex)
plt.show()
```
